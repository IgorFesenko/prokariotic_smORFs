# Parsing CD-HIT outup

```python
import pandas as pd
import numpy as np
import seaborn as sns
from matplotlib import pyplot as plt
```

# Erwinia

## Load clst.text

```python
clstr = pd.read_csv(r'../raw_data/erwinia_igs_orfs.pep_clstr_text', sep='\t')
```

```python
clstr.head()
```

```python
clstr.tail()
```

## Statistics

```python
clstr_above1 = clstr[clstr['clstr_size']>1]
```

```python
clstr_above1['clstr'].nunique()
```

```python
clstr_above1.head(10)
```

количесво кластеров размером больше 5

```python
clstr[clstr['clstr_size']>4]['clstr'].nunique()
```

```python
clstr[clstr['clstr_size']>4].head(10)
```

### Clusters from different genomes

```python
clstr_above1['Genome'] = clstr_above1['id'].map(lambda x: x.split(':')[2])
```

```python
clstr_above1.head()
```

Группируем по кластерам, чтобы посмотреть на количество геномов

```python
clstr_groupby = clstr_above1.groupby('clstr', as_index=False).agg({"Genome":list})
```

```python
clstr_groupby['Num_Genomes'] = clstr_groupby['Genome'].map(lambda x: len(x))
```

```python
clstr_groupby['Red_genomes'] = clstr_groupby['Genome'].map(lambda x: len(set(x)))
```

```python
clstr_groupby.head(10)
```

```python
clstr_groupby[clstr_groupby['Red_genomes']>1]['clstr'].nunique()
```

Количество кластеров без паралогов в геноме

```python
clstr_groupby['diff'] = clstr_groupby['Num_Genomes'] - clstr_groupby['Red_genomes']
```

```python
clstr_groupby_nonred = clstr_groupby[clstr_groupby['diff']==0]
```

```python
clstr_groupby_nonred[clstr_groupby_nonred['Red_genomes']>1]['clstr'].nunique()
```

```python
clstr_groupby_genomes = clstr_groupby.groupby('Num_Genomes', as_index=False).agg({'clstr':'count'})
clstr_groupby_genomes.nlargest(10, 'clstr')
```

```python
plt.figure(figsize=(15,8))
sns.set(style='white', font_scale=2)
sns.barplot(data=clstr_groupby_genomes.nlargest(10, 'clstr'), x='Num_Genomes', y='clstr')
plt.xlabel("The number of sequences in cluster", fontsize=24)
plt.ylabel("The number of clusters", fontsize=24)
plt.title("The top10 clusters", fontsize=30);
```

## Count different sequences

```python
pep_seq = pd.read_csv(r'../raw_data/erwinia_igs_orfs.pep.fa.gz.csv', compression='gzip')
```

```python
pep_seq.head()
```

```python
clstr_above1.head()
```

Сливаем сиквенсы с кластерами и группируем

```python
clstr_above1_seq = clstr_above1.merge(pep_seq[['ID','Sequence']], left_on='id', right_on='ID', how='left')
```

```python
clstr_above1_seq.head()
```

```python
clstr_above1_seq.info()
```

```python
clstr_groupby_seq = clstr_above1_seq.groupby('clstr', as_index=False).agg({"Sequence":list, "Genome":'count'})
```

```python
clstr_groupby_seq['iniq_seq'] = clstr_groupby_seq['Sequence'].map(lambda x: len(set(x)))
```

```python
clstr_groupby_seq.head()
```

```python
list(clstr_groupby_seq[clstr_groupby_seq['clstr']==3]['Sequence'])
```

Разница буквально в несколько аминокислот

Визуализируем распределение:

```python
clstr_groupby_seq[clstr_groupby_seq['iniq_seq']>4]['clstr'].nunique()
```

```python
clstr_groupby_seq[clstr_groupby_seq['iniq_seq']>7]['clstr'].nunique()
```

```python
plt.figure(figsize=(15,12))
sns.set(style='white', font_scale=2)
sns.barplot(data=clstr_groupby_seq.groupby('iniq_seq', as_index=False).agg({"clstr":'nunique'}),
                                           x='iniq_seq', y='clstr')
plt.xlabel("The number of unique sequences", fontsize=24)
plt.ylabel("The number of clusters", fontsize=24)
plt.title("The number of clusters", fontsize=30);
```

```python
clstr_groupby_seq[clstr_groupby_seq['iniq_seq']>7].head(30)
```

```python
list(clstr_groupby_seq[clstr_groupby_seq['clstr']==5884]['Sequence'])
```
